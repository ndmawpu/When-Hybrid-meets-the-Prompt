```python
from Prompt import prompt
import pandas as pd
import numpy as np
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df_anime = pd.read_csv("../Data/anime_adj.csv")
```

```python
class RecommenderPrompt:

    def __init__(self, df, vectors, input_prompt):
        self.df = df
        self.vectors = vectors
        self.input_prompt = input_prompt
        self.model = SentenceTransformer('all-MiniLM-L6-v2')
        self.rec_rs = None

    def userInput(self):
        user_input = self.input_prompt
        input_embeddings = self.model.encode(user_input)
        similarity = cosine_similarity(input_embeddings.reshape(1,input_embeddings.shape[0]),self.vectors)
        return similarity

    def recAnime(self):
        similarity_score = self.userInput()
        # sorting by similarity score
        if self.rec_rs is not np.nan:
            self.rec_rs = sorted(list(enumerate(similarity_score[0])), reverse=True, key=lambda x: x[1])
            print(self.rec_rs)
            return self.rec_rs

        # time to display best match according to prompt
        # if st.button('Recommend'):
        #     Model.ultils.print_rec(top_k=len(distances[0:15]),df=df_anime)
            # for i in distances[0:15]:
            #     try:
            #         rec_title = df_anime.iloc[i[0]]["item_name"]
            #         rec_avg_rating = df_anime.iloc[i[0]]["item_avg_rating"]
            #         rec_genre = df_anime.iloc[i[0]]["item_genre"]

            #         # divide the page into 2 columns
            #         col1, col2 = st.columns(2)

            #         # print the attributes on the page
            #         with col1:
            #             st.write(rec_title)
            #         with col2:
            #             st.write(rec_avg_rating)
            #             st.write(rec_genre)
            #     except:
            #         continue
```
